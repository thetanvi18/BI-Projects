```python
#Load dataset
import pandas as pd
df = pd.read_csv('customer_shopping_behaviour.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#Summary statistics, including categorical data
df.describe(include='all')
```

```python
#Checking for missing values
df.isnull().sum()
```

```python
#Imputing rating with median for each category to prevent bias
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#Renaming columns to snake casing for standardization
df.rename(columns=lambda x: x.strip().lower().replace(' ', '_'), inplace=True)
df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'}, inplace=True)
df.columns
```

```python
#Create new column age_group
labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
df[['age', 'age_group']].head(10)
```

```python
#Create new column purchase_frequency_days
frequency_mappings = {'Weekly': 7, 'Bi-Weekly': 14, 'Fortnightly': 14, 'Monthly': 30, 'Quarterly': 90, 'Every 3 Months': 90, 'Annually': 365}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mappings)
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
#Since both columns are identical, we can drop one
df.drop(columns=['promo_code_used'], inplace=True)
```

```python
df.columns
```

```python
#Save cleaned dataset
df.to_csv("customer_retail_clean.csv", index=False)
```
